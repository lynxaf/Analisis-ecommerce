## Exploratory Data Analysis (EDA)

### Explore ...

```python
merged_df = pd.merge(order_items, products, on='product_id', how='left')
merged_df = pd.merge(merged_df, order, on='order_id', how='left')
```

```python
merged_df.head()
```

```python
# Produk yang paling digemari dan paling sedikit digemari
product_sales = merged_df.groupby('product_category_name')['order_id'].count().sort_values(ascending=False)
```

```python
print("Produk yang paling digemari:")
print(product_sales.head(10))
```

```python
print("\nProduk yang paling sedikit digemari:")
print(product_sales.tail(10))
```

```python
# Waktu penjualan tertinggi
merged_df['order_purchase_month'] = merged_df['order_purchase_timestamp'].dt.month
monthly_sales = merged_df.groupby('order_purchase_month')['order_id'].count()
```

```python
print("\nWaktu penjualan tertinggi:")
print(monthly_sales)
```

**Insight:**
- berdasarkan EDA yang dilakukan ditemukan bahwa produk yang paling banyak terjual adalah cama_mesa_banho sedangkan produk yang paling sedikit terjual adalah seguros_e_servicos    
- kemudian waktu penjualan tertinggi terjadi pada bulan 8 dengan total penjualan sebanyak 12158

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Produk yang paling digemari
plt.figure(figsize=(10, 5))
sns.barplot(x=product_sales.head(10).index, y=product_sales.head(10).values)
plt.title('10 Produk Paling Digemari')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=90)
plt.show()


```

```python
# Produk yang paling sedikit digemari
plt.figure(figsize=(10, 5))
sns.barplot(x=product_sales.tail(10).index, y=product_sales.tail(10).values)
plt.title('10 Produk Paling Sedikit Digemari')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=90)
plt.show()
```

### Pertanyaan 2:

```python
# Visualisasi waktu penjualan tertinggi
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.title('Tren Penjualan Bulanan')
plt.xticks(range(1, 13)) # Menampilkan label bulan dari 1-12
plt.show()
```

```python
# Menampilkan bulan dengan penjualan tertinggi
highest_sales_month = monthly_sales.idxmax()
print(f"\nBulan dengan penjualan tertinggi: {highest_sales_month}")
```
